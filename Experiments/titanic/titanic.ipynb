```python
import pandas as pd

# Import data
data = pd.read_csv("tested.csv")

# Check null values
data.isna().sum()
```

```python
data.head()
```

```python
# Remove unnecessary columns
data_cleaned = data.drop("PassengerId", axis=1)
data_cleaned
```

```python
data_cleaned.isna().sum()
```

```python
# Fill the missing values in Age column
data_cleaned.fillna({"Age":data_cleaned["Age"].mean()}, inplace=True)
data_cleaned
```

```python
data_cleaned.isna().sum()
```

```python
data_cleaned["Cabin"].tail(10)
```

```python
# Remove unnecessary columns
data_cleaned = data_cleaned.drop(["Cabin", "Name"], axis=1)
data_cleaned.head()
```

```python
data_cleaned.isna().sum()
```

```python
# Check which row has Nan value for Fare column
data_cleaned[data_cleaned["Fare"].isnull()]
```

```python
# Drop row from dataframe where Fare column has Nan value
data_cleaned.dropna(subset=["Fare"], inplace=True)
```

```python
data_cleaned.isna().sum()
```

```python
# Split data in x and y
from sklearn.model_selection import train_test_split

x = data_cleaned.drop("Survived", axis=1)
y = data_cleaned["Survived"]

# Split data in training and test sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
print(len(x_test), len(y_test))
```

```python
# Convert category data into numbers (one hot encoding on Sex and Embarked columns)

from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer

category_columns = ["Sex", "Embarked", "Ticket"]

one_hot = OneHotEncoder(handle_unknown='ignore')
transformer = ColumnTransformer([(
                                "one_hot",
                                one_hot,
                                category_columns)],
                               remainder="passthrough")

x_train_transform = transformer.fit_transform(x_train)
x_test_transform = transformer.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier()
clf.fit(x_train_transform, y_train)
```

```python
from sklearn.metrics import accuracy_score
clf.score(x_test_transform, y_test)
# y_pred = clf.predict(x_test_transform)
# print(y_pred)
# accuracy_score(y_pred, y_test)
# print(len(x_test_transform))
```
