# Two options for subplots

```python
# 1st Option

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
```

```python
ax1.plot(x, x+2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(butter_price.keys(), butter_price.values());
ax4.plot(np.random.randn(1000));
```

```python
plt.show()
```

```python
# 2nd Option

fig, ax = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

ax[0, 0].plot(x, x+2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(butter_price.keys(), butter_price.values());
ax[1, 1].hist(np.random.randn(1000));
```

# OO method of matplotlib (Object Oriented)

```python
heart_deases.head()
```

```python
over_50 = heart_deases[heart_deases["age"] > 50]
```

```python
over_50.head()
```

```python
len(over_50)
```

```python
# pyplot method
over_50.plot(x="age", y="chol", kind="scatter", c="target");
```

```python
plt.show()
```

```python
# OO method mixed with pyplot method

fig, ax = plt.subplots(figsize=(10,6));
```

```python
over_50.plot(x="age",
            y="chol",
            kind="scatter",
            c="target",
            ax=ax);
```

```python
plt.show()
```

```python
## OO method from scratch

fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

ax.legend(*scatter.legend_elements(), title="target");
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
plt.show()
```

```python
over_50.head()
```

```python
# subplots for age, chol, thalach

fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))

# plot data on ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"]);

# customize data
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="chol");

# Add Legend
ax0.legend(*scatter.legend_elements(), title="target");

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="-");



# Plot data on ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# customize data
ax1.set(title="Heart Disease and thalach Levels",
       xlabel="age",
       ylabel="thalach");

# Add legend
ax1.legend(*scatter1.legend_elements(), title="target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");


# Add title to the figure
fig.suptitle("Hearth Disease Data", fontsize=16, fontweight="bold");
```

```python
# Set the style
plt.style.use('seaborn-v0_8-whitegrid')

# OO method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="winter");  # this changes the color scheme

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="Age",
      ylabel="Cholesterol");

# Add a legend
ax.legend(*scatter.legend_elements(), title="target");

# Add a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
plt.show()
```

```python
# Customizing the Y nd X axis limitations


# subplots for age, chol, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                               ncols=1,
                               figsize=(10, 10))

# plot data on ax0
scatter = ax0.scatter(x=over_50["age"],
                     y=over_50["chol"],
                     c=over_50["target"],
                     cmap="winter");

# customize data
ax0.set(title="Heart Disease and Cholesterol Levels",
       xlabel="Age",
       ylabel="chol");

# Change the x-axis limit
ax0.set_xlim([50, 80])

# Add Legend
ax0.legend(*scatter.legend_elements(), title="target");

# Add a meanline
ax0.axhline(y=over_50["chol"].mean(),
           linestyle="-");



# Plot data on ax1
scatter1 = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="winter")

# customize data
ax1.set(title="Heart Disease and thalach Levels",
       xlabel="age",
       ylabel="thalach");

# Change the x-axis limit
ax1.set_xlim([50, 80])
ax1.set_ylim([60,200])

# Add legend
ax1.legend(*scatter1.legend_elements(), title="target");

# Add a meanline
ax1.axhline(y=over_50["thalach"].mean(),
           linestyle="--");


# Add title to the figure
fig.suptitle("Hearth Disease Data", fontsize=16, fontweight="bold");
```
